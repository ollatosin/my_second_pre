
## Exploratory Data Analysis (EDA)

- Exploratory data analysis (EDA) is used by data scientists to analyze and investigate data sets and summarize their main characteristics, usually by employing data visualization methods. 
- It helps determine how best to manipulate data sources to get the answers you need, making it easier for data scientists to discover patterns, spot anomalies, test a hypothesis, or check assumptions.

 

### Types of EDA

- Univariate 

> Stem and leaf plots, Violin plots, Histograms, Box plots

- Multivariate

> Scatter plot, Multivariate charts, Run chart, Bubble chart, Heat map

## Univariate Analysis

```python
# Create a function  that labels 'survived' column properly

def survive(x):
    if x == 1:
        return 'Survived'
    else:
        return 'Died'
    
df['Survived_label'] = df['Survived'].apply(survive)
df 
```

```python
# create a visual (chart) to show distribution of 'died' and 'survived'

plt.figure(figsize=(9,8))
plt.title('Passenger Survival on the Titanic')
sns.countplot(x='Survived_label', data=df)
plt.xlabel('Survival')
plt.ylabel('Count of Passengers')
plt.show()
```

```python
df['Survived_label'].value_counts()
```

# Survival Rate
- What percentage of persons died vs survived?

```python
survival_percentage = (df['Survived_label'].value_counts()/df.shape[0]*100).round(1)
survival_percentage
```

```python
survived_count = df.groupby('Survived_label')['Survived_label'].count()
survived_count
```

```python
# create a visual (chart) to show distribution of 'died' and 'survived' with labels

plt.figure(figsize=(9,8))
plt.title('Passenger Survival on the Titanic')
sns.countplot(x='Survived_label', data=df)
for i, value in enumerate(survived_count.values):
    plt.text(i, value-100, str(value), fontsize=12)


plt.xlabel('Survival')
plt.ylabel('Count of Passengers')
plt.show()
```

```python
# Create visualization with label (a better method)

fig, ax1 = plt.subplots(figsize = (10, 5))
viz = sns.countplot(ax=ax1, x="Survived_label", data=df)
viz.set_xticklabels(viz.get_xticklabels(), rotation=90)
for i in viz.patches:
    height = i.get_height()
    viz.text(i.get_x() + i.get_width()/2., height + 0.1, height, ha="center")
plt.title("Passengers survivial on the Titanic")
plt.ylabel("Count of passengers")
```

```python
# create a visual (chart) to show distribution of 'died' and 'survived'

plt.figure(figsize=(9,8))
plt.title('Passenger Gender on the Titanic')
sns.countplot(x='Sex', data=df)
plt.xlabel('Survival')
plt.ylabel('Count of Passengers')
plt.show()
```

```python
# For the other univariate analysis, we will use subplots

fig, axes = plt.subplots(2,4, figsize=(20,10))

# categorical variables
sns.countplot(x='Survived_label', data=df, ax=axes[0,0])
sns.countplot(x='Pclass', data=df, ax=axes[0,1])
sns.countplot(x='Sex', data=df, ax=axes[0,2])
sns.countplot(x='SibSp', data=df, ax=axes[0,3])
sns.countplot(x='Parch', data=df, ax=axes[1,0])
sns.countplot(x='Embarked', data=df, ax=axes[1,1])


# numerical data
sns.histplot(x='Age', data=df, ax=axes[1,2])
sns.histplot(x='Fare', data=df, ax=axes[1,3])
```

### Bivariate Analysis
- Comparing/Analyzing two variables/features
- Combine Categorical vs Categorical or Categorical vs Numerical or Numerical vs Numerical vs Numerical


```python
# What was the survival rate by gender?

plt.figure(figsize=(12,8))
plt.title('Passenger Survival on the Titanic')
sns.countplot(x='Survived_label', data=df, hue='Sex')
plt.xlabel('Survived')
plt.ylabel('Count of Passengers')
```

```python

# Did place of embarkment play a role in survival?

plt.figure(figsize=(9,7))
plt.title('Passenger Survival on the Titanic')
sns.countplot(x='Survived_label', data=df, hue='Embarked')
plt.xlabel('Survived')
plt.ylabel('Count of Passengers')
```

```python
# Did passenger class have a role to play in survival?

plt.figure(figsize=(9,7))
plt.title('Passenger Survival on the Titanic')
sns.countplot(x='Survived_label', data=df, hue='Pclass')
plt.xlabel('Survived')
plt.ylabel('Count of Passengers')
```

### Multivariate Analysis

- More than one variable

```python
# Survival rate by passenger class by age
sns.boxplot(x='Pclass', y='Age', hue='Survived_label', data=df)
```

```python
# Mean age of each passenger class

# Wht do you notice?

df.groupby('Pclass')['Age'].mean()
```

```python
# Survival rate by Passenger class and Gender
# What do you observe?

sns.catplot(col='Survived_label', x='Sex', hue='Pclass', data=df, kind='count')
```

```python
# Pairplot plots each variable against other variables

sns.pairplot(df)
```

```python
# correlation matrix - shows the relationship among variables
# values range between -1 and 1
# Values very close to -1 indicate strong negative relationship
# Values very close to 1 indicate strong positive relationship




corel = df.corr()
sns.heatmap(corel,annot=True)
```
